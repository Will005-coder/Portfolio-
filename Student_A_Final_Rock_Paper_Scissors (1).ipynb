```python
#Author - William Dakare

#use .lower and strip() to translate player input into 1, 2, or 3 to represent rock, paper, and scissors
#this will make the back end booleans easier to code and understand
import random as r
import statistics
from collections.abc import Iterable


playerChoice = "" #Stores players attack choice
playerChoiceInt = 1 #Index number paired with an attack choice
indexOffset = 1
playerThrowRecord = []
i = [] #Track AI's history


options = {     #Dict that maps user input to an object: 1 maps to "rock" and so on
    "NUMBER": [1, 2, 3],
    "CHOICE": ["rock", "paper", "scissors"]
}


analysisDict = {     #Tracking AI success rate for each pattern being tracked. Modified in 'patterntrackingai'
                     #The naming for these nested dictionaries are explained in 'patterntrackingai'
      "Win": {
          "counterPreviousMove": 0, #Check if player counters their own move
          "counterAiPrevMove": 0, #Check if player counters the ai's move
      },
      "Loss": {
          "counterPreviousMove": 0,
          "counterAiPrevMove": 0,
      },
      "Tie": {
          "counterPreviousMove": 0,
          "repetition": 0,
          "oddMove": 0, #Next move doesn't counter the AI's move
      }
  }


successRate = {
      "Win": {
          "counterPreviousMove": {"timesPatternUSed": 0, "timesSuccessful": 0, "successPercent": 0 },
          "counterAiPrevMove": {"timesPatternUSed": 0, "timesSuccessful": 0, "successPercent": 0 },
      },
      "Loss": {
          "counterPreviousMove": {"timesPatternUSed": 0, "timesSuccessful": 0, "successPercent": 0 },
          "counterAiPrevMove": {"timesPatternUSed": 0, "timesSuccessful": 0, "successPercent": 0 },
      },
      "Tie": {
          "counterPreviousMove": {"timesPatternUSed": 0, "timesSuccessful": 0, "successPercent": 0 },
          "repetition": {"timesPatternUSed": 0, "timesSuccessful": 0, "successPercent": 0 },
          "oddMove": {"timesPatternUSed": 0, "timesSuccessful": 0, "successPercent": 0 },
    }}

aiPrevMove = None #Track's the AI's previous move


#------------Game Set-up/User Input---------------
def gamesetup():
  """Does game setup by printing:
     - game instructions and expected user input format,
     - option-value pairs for rock-paper-scissors mapping from 1 to 3.
     - The numbers correlate to the expected inputs. (1,2,3)==(rock,paper,scissors)
     None return type.
  """
  print("HELLO PLAYER 001")
  print("Welcome to Our Rock-Paper-Scissors game")
  print("Please enter the your attack choice when prompted to.")
  print("\nLET'S GO\n")

def validateinput(input: str, inputtype, bounds=("rock", "paper", "scissors"))-> bool:
  """
  Takes in input from the user, to check if it is in the expected format, and is either rock, paper or scissors.
  Args: Description
   input: Takes any form of string input.
   inputtype: Expected input format: string. [There's currently only need to validate inputs in str formats]
             If string expected, inputtype = str
  Example:
   input = "c", inputtype = str, bounds= ("a", "b", "c", "d"). Since "a" is a word, and is present in the bounds True(bool) is returned.
   input = "3e", inputtype = str, bounds(default). "3e" isn't a word. It doesn't check if it is in the bounds. False(bool) is returned.
  We check if input is in the format of a certain type, and not necessarily that type.
  """
  if isinstance(inputtype, str):
    isWord = input.isalpha()
    if isWord:
        #We have to make sure the input is either rock, paper or scissors.
      isValidInput = input in bounds
    else:
      isValidInput = False
  return isValidInput

def promptinput():
  """
  Prompts user for input based on the pairs printed.
  Conducts input validation.
  None return type.
  """
  isInputWord = False #User input seqeunce. Integrated with Validation.
  prefix = "" #This variable is a placeholder for a word added to the front of the input line when the user enters it in the wrong format.
  while not isInputWord:
    printstatement = f"{prefix}Your throw: "
    choice = input(printstatement).lower()
    isInputWord = validateinput(choice, str) #We expect the format of input to be in Alpha format and the variable will be returned as an (int) if its a word.
    if not isInputWord:
      prefix = "Seriously, "

  playerThrowRecord.append(choice)

def reset(): #Empty the pattern tracking dictionaries. This is done after each series.
    """
    Empties dicts `analysisDict` and `successRate` at the end of each series.
    """
    global analysisDict
    global successRate
    global playerThrowRecord # Reset player history
    global aiHistory # Reset AI history
    global aiPrevMove # Reset aiPrevMove

    playerThrowRecord = []
    aiHistory = []
    aiPrevMove = None

    for key in analysisDict.keys():
      for subKey in analysisDict[key].keys():
        analysisDict[key][subKey] = 0

    for key in successRate.keys():
      for subkey in successRate[key].keys():
        for categories in successRate[key][subkey].keys():
          successRate[key][subkey][categories] = 0


# AI Strategy functions
#---------------Random AI------------------------
def randomchoiceai() -> int:
    """
    Uses random function to generate choice between 1 and 3.
    1, 2, 3 correlates to rock-paper-scissors respectively.
    Returns:
        int: Random choice of 1 (rock), 2 (paper), or 3 (scissors)
    """
    aiThrow = r.randint(1, 3)
    return aiThrow


#-----------------Most Common and Pattern Tracking AI, shared functions--------------------
def decidewinner(aiThrow: int, playerThrow: int) -> str:
    """
    Determine winner of a round.
    Args:
        aiThrow (int): AI's throw (1=rock,2=paper,3=scissors)
        playerThrow (int): Player's throw (1=rock,2=paper,3=scissors)
    Returns:
        str: 'ai' if AI wins, 'player' if player wins, 'tie' if tie
    """
    assert isinstance(aiThrow, int) and 1 <= aiThrow <= 3
    assert isinstance(playerThrow, int) and 1 <= playerThrow <= 3
    if aiThrow == playerThrow:
        return "tie"

    elif (playerThrow - aiThrow) % 3 == 1:
        return "player"

    return "ai"


def gamelengthorfive(numOfRounds)-> int:
    """
    Returns the Number of matches played in the current series if <5, else 5.
    Args:
        numOfRounds (int): Total number of rounds played so far
    Returns:
        int: min(numOfRounds, 5)
    """
    assert isinstance(numOfRounds, int) and numOfRounds >= 0
    return min(numOfRounds, 5)



def findkeys(value, resultType, dictKeyGroup=None):
    """
    Return the second-level key(s) under `resultType` whose dictionary
    contains `value`.
    Parameters
    value : int | str | list
        Item to look for inside the pattern dictionaries.
    resultType : "Win" | "Loss" | "Tie"
        Top-level branch of `successRate` to search.
    dictKeyGroup : iterable | single value
        2nd-level branch. Candidate pattern name(s).  Each candidate is tested to find the key that matches `value` in `successRate`
    Returns
    str | int | list | None
        *Single value*  → first match or None.
        *Iterable input* → list of matches (possibly empty).
    """

    assert resultType in successRate
    dictType = type({})
    assert isinstance(resultType, (int, str, dictType))
    matches = [] #store multiple 'key' match


    for subCategory in dictKeyGroup:
        vals = successRate[resultType][subCategory]
        if value in vals.values() or value in vals:
            matches.append(subCategory)
        else:
            if value == vals:
                matches.append(subCategory)
    if isinstance(dictKeyGroup, list) and len(dictKeyGroup) == 1:
        return matches[0] if matches else None
    return matches #Single value 'dictKeyGroup'

#-------------------Most Common Throw AI-----------------
def mostcommoncounterai(playerThrowRecord, indexOffset)-> int:
  """
  Counter's the players most common choice, in their last 5 throws, or throughout the game.
  Returns int that represents the value of the AI's choice
  - 1 is rock
  - 2 is paper
  - 3 is scissors
  This is used in subsequent processing.

  Args:
  - playerThrowRecord
  - indexOffset(int): Pythons list positional indexing offset[pst: 1-> index: 0]
  """
  numOfRounds = len(playerThrowRecord) # Use actual number of rounds
  nOrFive = gamelengthorfive(numOfRounds)
  if nOrFive == 0: # Handle case with no history
      return randomchoiceai()
  last5orNMoves = playerThrowRecord[-nOrFive:] #Take the last 5 or N(game length) most current moves

  # Convert string choices to integers for statistics.mode
  last5orNMoves_int = [options["CHOICE"].index(move) + 1 for move in last5orNMoves]

  mostCommonLast5 = statistics.mode(last5orNMoves_int)


  #For the counter throw logic we'll use pick the next choice.
  #If the most common throw is rock (1), we pick paper(2).
  #The dictionary we're referencing is ordered in the form that the preceeding number defeats the previous one. And the 1st defeats the last.
  #To get the logic to work we can use +1 but that alone doesn't account for the counter to scissors(3) which is rock(1).
  #Looping around the list, however, can be done with the mod operation.
  endItem = 3
  if mostCommonLast5 == endItem: #if 3 we set back to 1
    counterChoice = 1
  else: #If I pick 1, we set the counter to choice 2
    counterChoice = mostCommonLast5 + 1

  #Python list indexing which we'll use next starts from 0 not 1. So we add 1 to counterChoice for indexing
  return counterChoice #We return counter, the int, because the choice-numeric-pair is used in deciding the winner of each round


#------------------PATTERN AI modules-------------------

def analyzepattern(resultKey, moveAtInstance, nextMoveInSeq):
    """
    Analyzes the player's decision pattern following a game outcome.
    This function updates the `analysisDict` based on how the player's next move
    relates to their previous move and the previous AI move.
    Args:
        resultKey (str): "Win", "Loss", or "Tie" based on the *previous* round outcome
        moveAtInstance (int): Player's move in the *previous* round (1=rock,2=paper,3=scissors)
        nextMoveInSeq (int): Player's move in the current round (1=rock,2=paper,3=scissors)
    """
    global aiPrevMove # Access the global aiPrevMove
    assert resultKey in analysisDict
    assert isinstance(moveAtInstance, int) and 1 <= moveAtInstance <= 3
    assert isinstance(nextMoveInSeq, int) and 1 <= nextMoveInSeq <= 3

    #Check if the player counters their own previous move
    isCounterOwnMove = (nextMoveInSeq == (moveAtInstance % 3) + 1)

    #Is the next move the counter to the AI's previous move
    isCounterAiMove = False
    if aiPrevMove is not None:
        isCounterAiMove = (nextMoveInSeq == (aiPrevMove % 3) + 1)

    #Tie has additional sub categories
    isTie = (resultKey == "Tie")
    #Move maps to an odd throw
    isRep = (moveAtInstance == nextMoveInSeq)
    isOdd = (nextMoveInSeq % 2 == 1)


    if isCounterOwnMove:
        analysisDict[resultKey]["counterPreviousMove"] += 1
    # Only consider counterAiPrevMove if it's not a tie, as tie has specific patterns
    elif isCounterAiMove and not isTie:
         analysisDict[resultKey]["counterAiPrevMove"] += 1
    elif isRep and isTie:
        analysisDict[resultKey]["repetition"] += 1
    elif isOdd and isTie:
        analysisDict[resultKey]["oddMove"] += 1

def responseanalysis(previousPlayerMove, currentPlayerMove, result):
    """"
    Checks for relationship between [currentPlayerMove] and [previousPlayerMove].
    Updates analysisDict based on result of previous round.
    Args:
        previousPlayerMove (int): Player's move in the previous round (1=rock,2=paper,3=scissors)
        currentPlayerMove (int): Player's move in the current round (1=rock,2=paper,3=scissors)
        result (str): Previous round outcome from player perspective ('ai','player','tie')
    Returns:
        dict: Updated analysisDict
    """
    assert result in ("ai", "player", "tie")
    if result == "ai":
        analyzepattern("Loss", previousPlayerMove, currentPlayerMove)
    elif result == "player":
        analyzepattern("Win", previousPlayerMove, currentPlayerMove)
    elif result == "tie":
        analyzepattern("Tie", previousPlayerMove, currentPlayerMove)
    return analysisDict


def updateanalysisfromrecords(playerMoves, aiMoves):
    """
    Updates analysisDict based on sequential history of moves.
    Args:
        playerMoves (list[int]): Chronological list of player moves (1=rock,2=paper,3=scissors)
        aiMoves (list[int]): Chronological list of AI moves (1=rock,2=paper,3=scissors)
    Returns:
        dict: Updated analysisDict
    """
    global aiPrevMove
    assert isinstance(playerMoves, list) and isinstance(aiMoves, list)
    n = len(playerMoves)
    if n < 2:
        return analysisDict

    # Iterate from the second move to analyze patterns based on the previous round
    # Convert string choices to integers for analysis
    playerMoves_int = [options["CHOICE"].index(move) + 1 for move in playerMoves]
    aiMoves_int = [options["CHOICE"].index(move) + 1 for move in aiMoves]


    for i in range(1, n):
        playerMove = playerMoves_int[i]
        previousPlayerMove = playerMoves_int[i - 1]
        # Get the AI move from the *previous* round
        aiPrevMove = aiMoves_int[i - 1] if i - 1 < len(aiMoves_int) else None

        if aiPrevMove is not None:
            # Determine the result of the *previous* round
            previousResult = decidewinner(aiPrevMove, previousPlayerMove)
            responseanalysis(previousPlayerMove, playerMove, previousResult)

    return analysisDict


def choosepatternfromdict(lastThrowResult):
  """
  Makes a subset of lists from `analysisDict` based on the result of the most recent throw: (WIN/LOSS/TIE)
  Args:
   - lastThrowResult(str): Result of the last throw
  Returns predicted player pattern for next throw.
  """
  if lastThrowResult not in analysisDict: # Added check for valid key
      return None
  multipleOccurences = [] #List that stores the patterns of instances where multiple patterns occur as frequently
  #First subset categorized by previous match result
  firstOrderSubSet = analysisDict[lastThrowResult]

  if not firstOrderSubSet or max(firstOrderSubSet.values()) == 0: #Checks for empty or all zero frequencies
      return None

  maxFrequency = max(firstOrderSubSet.values())
  for key, values in firstOrderSubSet.items():
    if maxFrequency == values: #Check if there are multiple instances with the same frequency
      multipleOccurences.append(key)


  if len(multipleOccurences) == 0:
      return None
  if len(multipleOccurences) == 1:
      dominantPattern = multipleOccurences[0]
      return dominantPattern


  # when multiple patterns tie, compare success rates
  bestPattern = None
  bestSuccess = 33 #Only take the success rates higher than 33%(That's the random probability rate, so this should be higher)
  for keyName in multipleOccurences:
      # Added checks for nested keys
      if lastThrowResult in successRate and keyName in successRate[lastThrowResult] and "successPercent" in successRate[lastThrowResult][keyName]:
          patternApplicationSuccess = successRate[lastThrowResult][keyName]["successPercent"]
          if patternApplicationSuccess > bestSuccess:
              bestSuccess = patternApplicationSuccess
              bestPattern = keyName

  if bestPattern is None: #if none of the multiple occurences have a good success rate
      bestPattern = r.choice(multipleOccurences)
  return bestPattern

def patterntrackingai(playerThrowRecord, aiHistory):
    """
    AI selects its next move based on player patterns.
    Uses `analysisDict` to determine which pattern is noticed:
    - Does the player throw the counter to thier previous move?
    - Does the player play the counter to the ai's move?
    - Falls back to random choice if no pattern is detectable

    SPECIAL CASE: If last throw is a tie
    - Check for repetitions of the same throw
    - Odd move on a tie
    Args:
        playerThrowRecord (list[str]): Historical player moves ("rock","paper","scissors")
        aiHistory (list[str]): Historical AI moves ("rock","paper","scissors")
    Returns:
        int: AI's next throw (1=rock,2=paper,3=scissors)
    Integration:
        pair with aithrowprint ELSE NO AI CHOICE
    """
    global aiPrevMove
    aiPrevMove = options["CHOICE"].index(aiHistory[-1]) + 1 if aiHistory else None # Convert to int

    updateanalysisfromrecords(playerThrowRecord, aiHistory)

    #Decide AI move based on last match result
    if len(playerThrowRecord) < 1 or len(aiHistory) < 1:
         if not playerThrowRecord:
             return randomchoiceai()
         else:
             matchResult = "tie" #Assume tie for the first round for pattern analysis
    else:
        lastPlayerMoveInt = options["CHOICE"].index(playerThrowRecord[-1]) + 1
        lastAiMoveInt = options["CHOICE"].index(aiHistory[-1]) + 1
        matchResult = decidewinner(lastAiMoveInt, lastPlayerMoveInt)


    #Choose the dominant pattern based on the last match result
    dominantPattern = choosepatternfromdict(matchResult)

    aiChoice = None # Initialize aiChoice

    #Based on the dominant pattern, determine the AI's next move
    if dominantPattern == "counterPreviousMove":
        if len(playerThrowRecord) >= 2:
            previousPlayerMoveInt = options["CHOICE"].index(playerThrowRecord[-2]) + 1
            predictedPlayerMoveInt = (previousPlayerMoveInt % 3) + 1 #Predicted player counter
            aiChoice = (predictedPlayerMoveInt % 3) + 1 #AI counters the predicted player counter
        else:
            #Not enough history for this pattern, fall back
             aiChoice = mostcommoncounterai(playerThrowRecord, indexOffset)

    elif dominantPattern == "counterAiPrevMove":
        if len(aiHistory) >= 1:
            lastAiMoveInt = options["CHOICE"].index(aiHistory[-1]) + 1
            predictedPlayerMoveInt = (lastAiMoveInt % 3) + 1 # Predicted player counter
            aiChoice = (predictedPlayerMoveInt % 3) + 1 # AI counters the predicted player counter
        else:
            # Not enough history for this pattern, fall back
            aiChoice = mostcommoncounterai(playerThrowRecord, indexOffset)

    elif dominantPattern == "repetition" and matchResult == "tie":
         if len(playerThrowRecord) >= 1:
            lastPlayerMoveInt = options["CHOICE"].index(playerThrowRecord[-1]) + 1
            aiChoice = (lastPlayerMoveInt % 3) + 1 # AI counters the repeated move
         else:
            aiChoice = mostcommoncounterai(playerThrowRecord, indexOffset)

    elif dominantPattern == "oddMove" and matchResult == "tie":
        # Fallback for less specific tie pattern
        aiChoice = mostcommoncounterai(playerThrowRecord, indexOffset)

    else:
        # If no specific dominant pattern or pattern not handled, fall back to most common counter.
        aiChoice = mostcommoncounterai(playerThrowRecord, indexOffset)

    if aiChoice is None:
        aiChoice = randomchoiceai()

    return aiChoice

def aithrowprint(aiThrowChoice: int):
    """
    Takes integer from AI strategies and prints the choice as the AI throw.
    Args:
        aiThrowChoice (int): AI's throw number (1=rock,2=paper,3=scissors)
    None Return type
    """
    assert isinstance(aiThrowChoice, int)
    aiThrowChoiceIndex = aiThrowChoice - indexOffset
    aiThrow = options["CHOICE"][aiThrowChoiceIndex]
    print("AI choice: ", aiThrow.title())

def selectAi(strat, playerHistory, aiHistory):
    """
    strat: 1 -> random, 2 -> most common, 3 -> pattern tracking
    """
    global indexOffset
    assert strat in (1, 2, 3)
    if strat == 1:
        return randomchoiceai()
    elif strat == 2:
        return mostcommoncounterai(playerHistory, indexOffset)
    return patterntrackingai(playerHistory, aiHistory)

# Demo / Example run

def randpercent(): #Make random successrates
    """
    Inserts random success percentages for each strategy `successRate` for testing
    """
    global successRate
    for key in successRate.keys():
      for subkey in successRate[key].keys():
        for categories in successRate[key][subkey].keys():
          successRate[key][subkey][categories] = r.randint(1, 100)

def runtest(): #call when testing
  """
  Runs 5 test cases for on each ai strategy.
  """
  randpercent()
  testCases = [
        {"name": "uniformRandom", "player": [r.choice(["rock", "paper", "scissors"]) for _ in range(10)]},
        {"name": "alwaysRock", "player": ["rock"] * 10},
        {"name": "cycleReps", "player": ["paper", "scissors", "rock", "paper", "scissors", "rock", "paper", "scissors", "rock", "paper"]},
        {"name": "repeatThenSwitch", "player": ["rock", "rock", "paper", "rock", "paper", "rock", "rock", "scissors", "rock", "rock"]},
        {"name": "shortHistory", "player": ["paper"]},
    ]
  for strat in (1, 2, 3):
    print(f"\n=== AI MODE {strat} ===")
    i = 1
    for key in testCases:
            reset() # Reset history and analysis for each test case
            playerThrowRecord.extend(key["player"]) # Use extend to add elements

            # Generate  AI history (random for testing)
            aiHistoryTest = [r.choice(["rock", "paper", "scissors"]) for i in range(len(playerThrowRecord))]

            print(f"\n--- Test case {i}: {key['name']} ---")

            print("Player history:", playerThrowRecord)
            print("AI history    :", aiHistoryTest)

            # set aiHistory for updateanalysisfromrecords
            global aiHistory
            aiHistory = aiHistoryTest.copy()

            # Run updateanalysisfromrecords with the test history
            updateanalysisfromrecords(playerThrowRecord, aiHistory)

            # Select AI move using the pattern tracking AI
            aiChoice = selectAi(strat, playerThrowRecord, aiHistory)

            print("Chosen next AI move (numeric):", aiChoice)
            print(analysisDict)
            aithrowprint(aiChoice)
            i += 1

aiPrevMove = None
runtest()
```
