Option 1: Single Line Deep Dive
Analyze one MBTA line (Red, Orange, **Green (B,** C, D, or E), Blue, or a bus route) in detail.
Requirements:
- Fetch real-time **prediction** **data for all stops** on your chosen line
- Calculate **average wait times at each station**
- Identify the busiest stations (most frequent service)**[mode of overall data]**
- *Compare wait times in different directions (inbound vs outbound)*
- Use numpy to calculate: mean, median, standard deviation of wait times
Deliverables:
- **Summary** statistics table (pandas DataFrame)
- *Visualization **comparing wait times** across stations*
- **Written analysis** identifying patterns or **insights**

```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # for visualizations

stations = ['South Station', 'Park Street', 'Downtown Crossing', 'Kendall/MIT', 'Central']
wait_times = [4.2, 3.8, 5.1, 4.5, 3.9]  # in minutes

# Bar chart for
plt.figure(figsize=(10, 6))  # Set the size of the figure
plt.bar(stations, wait_times, color='red', alpha=0.7)  # alpha controls transparency
plt.xlabel('Station', fontsize=12)
plt.ylabel('Average Wait Time (minutes)', fontsize=12)
plt.title('Average Wait Times at Red Line Stations', fontsize=14, fontweight='bold')
plt.xticks(rotation=15, ha='right')  # Rotate station names for readability
plt.legend(fontsize=11)  # Show the legend
plt.grid(True, color='g', linestyle='-', linewidth=2, alpha=0.2)
plt.tight_layout()  # Prevents labels from being cut off
plt.savefig('wait_times_bar.png')  # Save the figure
plt.show()
```

```python
lines = ['Red Line', 'Orange Line', 'Green Line', 'Blue Line']
avg_waits = [4.3, 5.2, 6.1, 4.8]

plt.figure(figsize=(8, 6))
plt.barh(lines, avg_waits, color=['red', 'orange', 'green', 'blue'])
plt.xlabel('Average Wait Time (minutes)', fontsize=12)
plt.ylabel('Transit Line', fontsize=12)
plt.title('Comparison of Wait Times Across MBTA Lines', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('line_comparison.png')
plt.show()
```
